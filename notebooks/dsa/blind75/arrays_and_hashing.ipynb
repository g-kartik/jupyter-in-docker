# Contains duplicate
https://leetcode.com/problems/contains-duplicate/

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

Constraints:

- 1 <= nums.length <= 105
- -10^9 <= nums[i] <= 10^9

## Pseudocode
### Bruteforce

```
def containsDuplicate(self, nums: List[int]) -> bool:
    for i from 0 to nums.length - 1:
        el1 = nums[i]
        for j from i + 1 to nums.length - 1:
            el2 = nums[j]
            if el1 == el2:
                return True
        return False
```

### Optimal
```

    map = dict()
    for n in nums:
        if n not in map:
            add n to map
        else:
            return True

## Takeaways
- The bruteforce method has O(n^2) worst case time complexity, while the optimal method has only O(n) worst case time complexity
- We can tradeoff space complexity for optimizing time complexity. 

# Valid Anagram
https://leetcode.com/problems/valid-anagram/

## Statement
Given two strings s and t, return true if t is an anagram of s, and false otherwise.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

Constraints:

- 1 <= s.length, t.length <= 5 * 104
- s and t consist of lowercase English letters.

Follow up: What if the inputs contain Unicode characters? How would you adapt your solution to such a case?

````python
def isAnagram_1(self, s: str, t: str) -> bool:
    if len(s) != len(t):
        return False
    map_s = {}
    map_t = {}
    for char1, char2 in zip(s, t):
        try:
            map_s[char1] += 1
        except KeyError:
            map_s[char1] = 1

        try:
            map_t[char2] += 1
        except KeyError:
            map_t[char2] = 1

    return map_s == map_t

def isAnagram(self, s: str, t: str) -> bool:
    if len(s) != len(t):
        return False
    map = {}
    for char1, char2 in zip(s, t):
        try:
            map[char1] += 1
        except KeyError:
            map[char1] = 1

        try:
            map[char2] -= 1
        except KeyError:
            map[char2] = -1

    return not any([v for v in map.values()])
````
