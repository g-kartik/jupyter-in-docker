## Naive approach
### Pseudocode

- max_product <-- 0
- Iterate through the list
- For each element, find its product with all the next elements in the list and if any product value is greater than the max_product, then update the max_product value
- return max_product

```python
def max_pairwise_product(array):
    n = len(array)
    max_product = 0
    for first in range(n):
        for second in range(first+1, n):
            max_product = max(max_product, array[first]*array[second])
    return max_product
```

## Efficient Approach
### Pseudocode

- Initialize two variables, first_largest and second largest to be -1 and None respectively
- Iterate through the array
- If the iterated element is greater than or equal to the first largest, copy the value of the first largest to the second largest and update the first largest value
- Return the product of first and second largest values

```python
def max_pairwise_product_2(array):
    first = -1
    second = -1
    
    for el in array:
        if el > first:
            second = first
            first = el
        elif el > second:
            second = el
    return first * second
```

## Challenge 1: Find two largest elements in an array in 2n comparisions
### Pseudocode
- Find the largest element in the array and remember its index
- Find the second largest element in the array excluding the first largest
- Return the product of first and second largest elements

```python
def largest_elements_2n(array):
    n = len(array)
    index = 0
    for i in range(1, n):
        if array[i] > array[index]:
            index = i
    array[index], array[n-1] = array[n-1], array[index]
    index = 0
    for i in range(1, n-1):
        if array[i] > array[index]:
            index = i
    array[index], array[n-2] = array[n-2], array[index]    
    return array[n-1], array[n-2]
```

## Challenge 2: Find two largest elements in an array in 1.5n comparisions
### Pseudocode

- Iterate through the array in pairs and swap the elements such that the left element is greater than the right
- Initialize two pointers, first and second to be zero
- Now iterate only through the even element positions of the array starting from the second even position
- If the element at the index is greater than the element at the first then copy the first to the second and copy the index to the first
- Elif the element at the index is greater than the element at the second then copy the index to the second
- Check if the element at (first + 1) is greater than or equal to  the element at second. If assign second as (first + 1)
- Return the product of elements at first and second.

```python
def max_pairwise_1_5n(array):
    
    for i in range(0, len(array)-1, 2):
        if array[i] < array[i+1]:
            array[i], array[i+1] = array[i+1], array[i]
    
    first = second = -1
    first_index = None
    
    for i in range(0, len(array), 2):
        el = array[i]
        
        if el > first:
            second = first
            first = el
            first_index = i
        elif el > second:
            second = el
    try:        
        if array[first_index+1] > second:
            second = array[first_index + 1]
    except IndexError:
        pass
        
    return first * second
```
